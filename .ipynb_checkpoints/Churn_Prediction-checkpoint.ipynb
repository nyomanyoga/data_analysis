```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

# Impor library untuk pemodelan
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Ganti 'data.csv' dengan path file csv kamu
customers = pd.read_csv('./customers.csv')
shipments = pd.read_csv('./shipments.csv')
transactions = pd.read_csv('./transactions.csv')
```

## Task 1: Build a machine learning model (Model Baseline)

```python
# 1. Konversi tipe data tanggal
customers['join_date'] = pd.to_datetime(customers['join_date'])
transactions['order_date'] = pd.to_datetime(transactions['order_date'])

# 2. Persiapan data awal
today_date = transactions['order_date'].max() + pd.Timedelta(days=1)
customer_summary = transactions.groupby('customer_id').agg(
    last_order_date=('order_date', 'max'),
    total_orders=('order_id', 'count'),
    avg_shipping_fee=('shipping_fee', 'mean')
).reset_index()
df_model = pd.merge(customers, customer_summary, on='customer_id', how='left')

# 3. Membuat variabel target 'is_churn'
df_model['days_since_last_order'] = (today_date - df_model['last_order_date']).dt.days
df_model['is_churn'] = (df_model['days_since_last_order'] >= 90).astype(int)

# Menangani nilai kosong setelah semua perhitungan
df_model.fillna(0, inplace=True)
```

```python
print("--- Memulai Task 1: Pelatihan Model Baseline ---")
# Mendefinisikan fitur dasar dan target
features_model_1 = ['total_orders', 'avg_shipping_fee']
target = 'is_churn'
y = df_model[target]

# Membagi data menjadi train dan test.
X_train, X_test, y_train, y_test = train_test_split(
    df_model, y, test_size=0.3, random_state=42
)

# Inisialisasi dan latih Model 1
model_1 = RandomForestClassifier(random_state=42)
model_1.fit(X_train[features_model_1], y_train)

# Simpan hasil prediksi untuk evaluasi akhir di Task 3
y1_pred = model_1.predict(X_test[features_model_1])
y1_proba = model_1.predict_proba(X_test[features_model_1])[:, 1]

print("--- Model 1 (RF Baseline) Selesai Dilatih ---")
```

## Task 2: Perform relevant feature engineering (Enhancement)

```python
print("--- Memulai Task 2: Feature Engineering dan Pelatihan Ulang ---")
# 1. Feature Engineering: Menambahkan fitur 'customer_age_days'
X_train['customer_age_days'] = (today_date - X_train['join_date']).dt.days
X_test['customer_age_days'] = (today_date - X_test['join_date']).dt.days
X_train.fillna(0, inplace=True)
X_test.fillna(0, inplace=True)
print("Fitur 'customer_age_days' telah ditambahkan.")

# 2. Mendefinisikan set fitur baru yang sudah diperkaya
features_model_2 = ['total_orders', 'avg_shipping_fee', 'customer_age_days']

# 3. Inisialisasi dan latih Model 2
model_2 = RandomForestClassifier(random_state=42)
model_2.fit(X_train[features_model_2], y_train)

# Simpan hasil prediksi untuk evaluasi akhir di Task 3
y2_pred = model_2.predict(X_test[features_model_2])
y2_proba = model_2.predict_proba(X_test[features_model_2])[:, 1]

print("--- Model 2 (RF Enhanced) Selesai Dilatih ---")
```

## Task 3: Explain evaluation metric & Build Comparison Model

```python
print("--- Memulai Task 3: Penanganan Imbalance dan Evaluasi Akhir ---")

# --- Membangun Model 3 (SMOTE) ---
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train[features_model_2], y_train)
model_3 = RandomForestClassifier(random_state=42)
model_3.fit(X_train_smote, y_train_smote)
y3_pred = model_3.predict(X_test[features_model_2])
y3_proba = model_3.predict_proba(X_test[features_model_2])[:, 1]
print("--- Model 3 (SMOTE) Selesai Dilatih ---")

# --- Membangun Model 4 (Random Under-sampling) ---
rus = RandomUnderSampler(random_state=42)
X_train_rus, y_train_rus = rus.fit_resample(X_train[features_model_2], y_train)
model_4 = RandomForestClassifier(random_state=42)
model_4.fit(X_train_rus, y_train_rus)
y4_pred = model_4.predict(X_test[features_model_2])
y4_proba = model_4.predict_proba(X_test[features_model_2])[:, 1]
print("--- Model 4 (Under-sampling) Selesai Dilatih ---")


# --- Evaluasi Akhir: Perbandingan Semua Model ---
print("\n\n" + "="*35)
print("HASIL AKHIR & PERBANDINGAN SEMUA MODEL")
print("="*35)

# Visualisasi Imbalance pada data asli dalam bentuk Pie Chart
churn_counts = y.value_counts()
plt.figure(figsize=(8, 6))
plt.pie(churn_counts, labels=['Churn (1)', 'Tidak Churn (0)'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Proporsi Data Churn vs Tidak Churn', fontsize=16)
plt.show()

# Hitung metrik untuk semua model
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

f1_1, roc_1 = f1_score(y_test, y1_pred, zero_division=0), roc_auc_score(y_test, y1_proba) if len(set(y_test)) > 1 else 0
precision_1, recall_1 = precision_score(y_test, y1_pred, zero_division=0), recall_score(y_test, y1_pred, zero_division=0)

f1_2, roc_2 = f1_score(y_test, y2_pred, zero_division=0), roc_auc_score(y_test, y2_proba) if len(set(y_test)) > 1 else 0
precision_2, recall_2 = precision_score(y_test, y2_pred, zero_division=0), recall_score(y_test, y2_pred, zero_division=0)

f1_3, roc_3 = f1_score(y_test, y3_pred, zero_division=0), roc_auc_score(y_test, y3_proba) if len(set(y_test)) > 1 else 0
precision_3, recall_3 = precision_score(y_test, y3_pred, zero_division=0), recall_score(y_test, y3_pred, zero_division=0)

f1_4, roc_4 = f1_score(y_test, y4_pred, zero_division=0), roc_auc_score(y_test, y4_proba) if len(set(y_test)) > 1 else 0
precision_4, recall_4 = precision_score(y_test, y4_pred, zero_division=0), recall_score(y_test, y4_pred, zero_division=0)

# Tampilkan tabel perbandingan
print("\n========================================== Perbandingan Performa Semua Model ===========================================")
print(f"| Metrik        | Model 1 (RF Baseline)    | Model 2 (RF Enhanced)     | Model 3 (SMOTE)  | Model 4 (Under-sampling) |")
print("----------------------------------------------------------------------------------------------------------------------")
print(f"| F1-Score      | {f1_1:.4f}                    | {f1_2:.4f}                    | {f1_3:.4f}          | {f1_4:.4f}                   |")
print(f"| Precision     | {precision_1:.4f}                    | {precision_2:.4f}                    | {precision_3:.4f}          | {precision_4:.4f}                   |")
print(f"| Recall        | {recall_1:.4f}                    | {recall_2:.4f}                    | {recall_3:.4f}          | {recall_4:.4f}                   |")
print(f"| ROC-AUC Score | {roc_1:.4f}                    | {roc_2:.4f}                    | {roc_3:.4f}          | {roc_4:.4f}                   |")
print("======================================================================================================================\n")
```

## Laporan Model Prediksi Churn
### Task 1: Data Preparation & Model Baseline

Tujuan dari *task* ini adalah menyiapkan data dan membangun model dasar sebagai acuan (*baseline*).

1.  **Data Preparation**
    * Menggabungkan data `customers` (informasi pelanggan) dan `transactions` (riwayat pesanan) untuk mendapatkan gambaran lengkap per pelanggan.
    * Mendefinisikan **churn**: Pelanggan dianggap *churn* (label = 1) jika tidak bertransaksi selama 90 hari atau lebih.

2.  **Feature Engineering (Dasar)**
    * `total_orders` → Jumlah total pesanan yang pernah dilakukan.
    * `avg_shipping_fee` → Rata-rata ongkos kirim dari semua pesanan.

3.  **Pembuatan Model Baseline (Model 1)**
    * Sebuah model **Random Forest** dilatih hanya dengan menggunakan dua fitur dasar di atas.
    * Tujuannya adalah untuk mendapatkan skor performa awal sebelum dilakukan optimasi lebih lanjut.

---
### Task 2: Feature Engineering & Peningkatan Model

Pada *task* ini, kita mencoba meningkatkan performa model dengan menambahkan fitur baru yang lebih deskriptif.

1.  **Feature Engineering (Tambahan)**
    * `customer_age_days` → Umur keanggotaan pelanggan dalam hari, dihitung dari tanggal bergabung hingga hari ini. Fitur ini memberikan konteks seberapa baru atau lama seorang pelanggan.

2.  **Peningkatan Model (Model 2)**
    * Model **Random Forest** baru dilatih ulang dengan menambahkan `customer_age_days` ke dalam set fitur.
    * Tujuannya adalah untuk melihat apakah informasi umur pelanggan dapat membantu model memprediksi *churn* dengan lebih baik.

---
### Task 3: Penanganan Imbalance Data & Perbandingan Akhir

Data *churn* seringkali tidak seimbang. *Task* ini fokus pada penanganan masalah tersebut dan membandingkan semua model yang telah dibuat.

1.  **Analisis Imbalance Data**
    * Visualisasi proporsi data menunjukkan bahwa jumlah pelanggan yang *churn* dan tidak *churn* tidak seimbang.

    **Imbalance Teknik Digunakan Untuk Melihat Perbandingan Performance**
    -   Model yang dilatih pada data tidak seimbang cenderung "malas" dan lebih sering menebak kelas mayoritas.
    -   Akibatnya, model gagal mengidentifikasi pelanggan yang akan *churn* (kelas minoritas), padahal itulah tujuan utama kita.

2.  **Penerapan Teknik Imbalance**
    * **SMOTE (Oversampling)** → Menciptakan data sintetis untuk kelas minoritas agar jumlahnya seimbang. **Model 3** dilatih dengan teknik ini.
    * **RandomUnderSampler** → Mengurangi data dari kelas mayoritas secara acak. **Model 4** dilatih dengan teknik ini.

---
### Kesimpulan & Next Steps
* **Model terbaik adalah Model 4** yang menggunakan Feature Engineering dan teknik **RandomUnderSampler** untuk menangani data tidak seimbang.
* **Feature engineering** saja tidak cukup jika masalah mendasarnya adalah data yang tidak seimbang.
* Penanganan *imbalance* adalah **langkah krusial** untuk mendapatkan model prediksi *churn* yang benar-benar berguna.
