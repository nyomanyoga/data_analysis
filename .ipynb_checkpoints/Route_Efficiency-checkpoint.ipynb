```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
```

```python
# Ganti 'data.csv' dengan path file csv kamu
customers = pd.read_csv('./customers.csv')
shipments = pd.read_csv('./shipments.csv')
transactions = pd.read_csv('./transactions.csv')
```

## Task 1: Identify the routes (origin â†’ destination) with the highest frequency of delays.

```python
# Langkah 1: Membuat kolom 'route' unik untuk setiap kombinasi asal dan tujuan.
shipments['route'] = shipments['origin_city'] + ' -> ' + shipments['destination_city']

# Langkah 2: Mengelompokkan data berdasarkan rute dan melakukan agregasi.
route_summary = shipments.groupby('route').agg(
    total_shipments=('shipment_id', 'count'),
    total_delays=('delay_flag', 'sum')
).reset_index()

# Langkah 3: Menghitung kolom 'delay_rate' dalam format persentase.
route_summary['delay_rate (%)'] = (route_summary['total_delays'] / route_summary['total_shipments']) * 100

# Langkah 4: Mengurutkan hasil untuk menampilkan rute paling bermasalah di atas.
top_delayed_routes = route_summary.sort_values(by='delay_rate (%)', ascending=False)

# Menampilkan hasil akhir
print("--- [HASIL TASK 1] Peringkat Rute dengan Frekuensi Keterlambatan Tertinggi ---")
top_delayed_routes
```

## Task 2: Perform clustering to group routes based on efficiency levels

```python
# --- Langkah 2.1: Feature Engineering di Tingkat Rute ---
if 'route' not in shipments.columns:
    shipments['route'] = shipments['origin_city'] + ' -> ' + shipments['destination_city']

route_features = shipments.groupby('route').agg(
    total_shipments=('shipment_id', 'count'),
    total_delays=('delay_flag', 'sum'),
    avg_delivery_duration=('delivery_duration', 'mean'),
    avg_expected_duration=('expected_duration', 'mean')
).reset_index()
route_features['delay_rate'] = route_features['total_delays'] / route_features['total_shipments']
route_features['avg_duration_diff'] = route_features['avg_delivery_duration'] - route_features['avg_expected_duration']
print("\nFitur untuk clustering berhasil dibuat.")

# --- Langkah 2.2: Penskalaan Fitur (Feature Scaling) ---
features_for_clustering = ['total_shipments', 'delay_rate', 'avg_delivery_duration', 'avg_duration_diff']
X = route_features[features_for_clustering]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Fitur berhasil diskalakan.")

# --- Langkah 2.3: Menentukan Jumlah Cluster Optimal (Metode Silhouette Score) ---
silhouette_scores = []
K = range(2, 8) # Silhouette score dihitung mulai dari k=2

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X_scaled)
    score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plotting Silhouette Scores
plt.figure(figsize=(8, 5))
plt.plot(K, silhouette_scores, 'bo-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.title('Metode Silhouette Score untuk Menentukan k Optimal')
plt.show()

# Menemukan k dengan skor tertinggi secara otomatis
optimal_k = K[silhouette_scores.index(max(silhouette_scores))]
print(f"\n--- [HASIL TASK 2] Jumlah cluster optimal berdasarkan Silhouette Score adalah: {optimal_k} ---")
```

## Task 3: Provide recommendations for the operations team based on your analysis

```python
# ======================================================================
# Bagian 1: Menerapkan Model Final dengan k optimsl
# ======================================================================
# Menerapkan algoritma K-Means final
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
kmeans_final.fit(X_scaled)

# Menambahkan label cluster ke DataFrame utama
route_features['cluster'] = kmeans_final.labels_

# ======================================================================
# Bagian 2: Analisis Kuantitatif & Visual
# ======================================================================
# Membuat tabel statistik yang menjadi dasar semua keputusan kita.
features_for_clustering = ['total_shipments', 'delay_rate', 'avg_delivery_duration', 'avg_duration_diff']
cluster_profile = route_features.groupby('cluster')[features_for_clustering].mean()

print("\n--- Tabel Perbandingan Karakteristik Rata-rata per Cluster ---")
cluster_profile
```

```python
# Visualisasi sebaran data untuk 2 cluster
print("\n--- Menampilkan Visualisasi Sebaran Cluster ---")
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=route_features, x='delay_rate', y='avg_duration_diff',
    hue='cluster', palette='viridis', s=150, alpha=0.9, style='cluster',
    edgecolor='w'
)
plt.title('Sebaran 2 Cluster Rute Berdasarkan Efisiensi', fontsize=16)
plt.xlabel('Tingkat Keterlambatan (Semakin ke kanan, semakin buruk)')
plt.ylabel('Selisih Durasi (Semakin ke atas, semakin lambat dari janji)')
plt.grid(True)
plt.axhline(0, color='grey', linestyle='--')
plt.legend(title='Cluster')
plt.show()
```

```python
# ======================================================================
# Bagian 3: Interpretasi Otomatis untuk Laporan Final
# ======================================================================

interpretations = []

# Secara statistik, kita tentukan cluster 'bermasalah' adalah yang
# memiliki rata-rata tingkat keterlambatan (delay_rate) tertinggi.
problem_cluster_index = cluster_profile['delay_rate'].idxmax()

for c, group_stats in cluster_profile.iterrows():
    
    if c == problem_cluster_index:
        archetype = "Rute Tidak Efisien / Berisiko"
        reason = "Secara statistik, cluster ini menunjukkan performa terendah dengan rata-rata tingkat keterlambatan dan selisih durasi yang signifikan lebih tinggi."
        rec = "Prioritaskan untuk investigasi mendalam; Lakukan audit operasional pada rute-rute utama di cluster ini; Tinjau ulang SLA."
    else:
        archetype = "Rute Efisien / Standar"
        reason = "Secara statistik, cluster ini merepresentasikan standar operasional yang lebih baik dengan tingkat keterlambatan yang jauh lebih rendah."
        rec = "Jadikan sebagai benchmark; Pertahankan stabilitas performa, terutama pada rute bervolume tinggi; Terapkan best practice dari rute ini."
    
    interpretations.append({
        "Cluster": c,
        "Persona": archetype,
        "Alasan Segmentasi": reason,
        "Rekomendasi Strategis": rec
    })

interpret_df = pd.DataFrame(interpretations).set_index('Cluster')

# ======================================================================
# HASIL AKHIR: Menampilkan Tabel Laporan
# ======================================================================
print("--- [LAPORAN FINAL] Tabel Interpretasi & Rekomendasi ---")

# Menampilkan DataFrame dengan style agar lebih mudah dibaca di notebook
display(interpret_df.style.set_properties(**{
    'text-align': 'left',
    'white-space': 'normal'
}).set_table_styles(
    [dict(selector='th', props=[('text-align', 'left')])]
))
```
